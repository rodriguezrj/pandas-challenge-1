## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

# New column 'subtotal'
df['subtotal'] = df['unit_price'] * df['qty']

# Print
print(df.head())
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calculate_shipping_price(unit_weight):
    if unit_weight > 50:
        return 7 * unit_weight
    else:
        return 10 * unit_weight
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(calculate_shipping_price)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)


```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

def calculate_total_price(subtotal, shipping_price, tax_rate):
    sales_tax = (subtotal + shipping_price) * (tax_rate / 100)
    total_price = subtotal + shipping_price + sales_tax
    return total_price

# Sales tax rate of 9.25%
sales_tax_rate = 9.25 / 100

# New column 'total_price'
df['total_price'] = df.apply(lambda row: calculate_total_price(row['subtotal'], row['shipping_price'], sales_tax_rate), axis=1)

df["total_price"] = df["total_price"].round(2)


# print
df[['shipping_price', 'subtotal', 'total_price']].head(3)

```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# New column 'line_cost'
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

# Print
print(df.head())
```

```python
# Create a column for the profit of
# each line using line cost and line price

# Line price
df['line_price'] = df['unit_price'] * df['qty']

# Line cost
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

# New column 'line_profit'
df['line_profit'] = df['line_price'] - df['line_cost']

# Print
print(df.head())
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

# Group  by client ID
client_spending = df.groupby('client_id').agg(
    total_spending=('subtotal', 'sum'),
    total_shipping=('shipping_price', 'sum')
)

# Total spending 
client_spending['total_spending_including_shipping'] = client_spending['total_spending'] + client_spending['total_shipping']

# Top 5 clients
top_clients_by_quantity = df['client_id'].value_counts().head(5)

top_clients_spending = client_spending.loc[top_clients_by_quantity.index]

print(top_clients_spending)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Group by client ID
client_summary = df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('subtotal', 'sum'),
    total_profit=('line_profit', 'sum')
)

# Top 5 clients
top_clients = client_summary.nlargest(5, 'total_profit')

# Print Summary
print(top_clients)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Calculate total revenue and total profit
df['total_revenue'] = df['subtotal'] + df['shipping_price']
df['total_profit'] = df['total_revenue'] - df['line_cost']

# Convert currency columns to millions of dollars
df['total_revenue_million'] = df['total_revenue'] / 1e6
df['total_profit_million'] = df['total_profit'] / 1e6

# Rename columns
df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price (Millions of Dollars)',
    'total_revenue_million': 'Total Revenue (Millions of Dollars)',
    'total_profit_million': 'Total Profit (Millions of Dollars)'
}, inplace=True)

formatted_df = df[['Client ID', 'Total Units Purchased', 'Total Shipping Price (Millions of Dollars)', 
                   'Total Revenue (Millions of Dollars)', 'Total Profit (Millions of Dollars)']]

# Print
print(formatted_df)



```
